```python
import epw_columns
import pandas as pd
import glob
epwFiles = glob.glob('./*.epw')
```

```python
mesFichiers = {}
for file in epwFiles:
    name = file.split('\\')[-1].split('.epw')[0]
    mesFichiers[name] = pd.read_csv(file, names = epw_columns.get(), skiprows = 8)
    try:mesFichiers[name].index = pd.date_range(start = '2021-01-01 01:00:00', end = '2022-01-01 00:00:00', freq = '1h')
    except:mesFichiers[name].index = pd.date_range(start = '2021-01-01 01:00:00', end = '2022-01-02 00:00:00', freq = '1h')       
```
